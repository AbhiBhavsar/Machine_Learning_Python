## Matplotlib workflow

```python
import matplotlib.pyplot as plt
#1. prepare data
x=[np.random.randint(0,5) for _ in range(5)]
y=[np.random.randint(2,7) for _ in range(5)]
print(x, y)

#2. Setup Plot
fig,ax= plt.subplots(figsize=(10,10)) #(height,width)

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Dummy plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save/Share
fig.savefig("imgs/dummyPlot.png")
```

## Subplots

```python
#1st method
x=np.random.random(5)
y=np.random.random(5)
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2, ncols=2, figsize=(10,5))
ax1.set(title="Line Plot")
ax1.plot(x,y)
ax2.set(title="Scatter Plot")
ax2.scatter(x,y)
ax3.set(title="bar Plot")
ax3.bar(x,y)
ax4.set(title="Hist Plot")
ax4.hist(x)
# we can put ; to hide python o/p
```

```python
# 2nd Way to create subplots
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,y)
ax[0,1].scatter(x,y)
```

## Ploting with Pandas D-Frame

```python
car_sales=pd.read_csv("car-sales.csv")
car_sales["Price"]=car_sales["Price"].str.replace('[$|.|,]','', regex=True).str[:-2]
car_sales['Sale Date']=pd.date_range('01-05-2024',periods=len(car_sales))
car_sales["Sales Overtime"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x='Sale Date',y='Sales Overtime',kind='bar')
```

## Working with OO methods

```python
patient_csv=pd.read_csv("heart-disease.csv")
patient_csv
```

```python
age_50_plus=patient_csv[patient_csv['age']>50] #boolean indexing
age_50_plus
```

```python
fig, ax=plt.subplots(figsize=(10,6))
age_50_plus.plot(kind='scatter',
                x='age',
                y='chol',
                c='target',
                ax=ax);
ax.set_xlim([45,95])
```

```python
## Only OO approach
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(x=age_50_plus['age'],
                  y=age_50_plus['chol'],
                  c=age_50_plus['target'])
ax.set(title="Patient 50+",xlabel='Age',ylabel='Cholestrol')
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(age_50_plus.chol.mean(),linestyle='-.')
```
